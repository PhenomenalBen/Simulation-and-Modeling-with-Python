# Exercise 6 - [10 pt]

So far we have considered two models of a falling penny:

* If we ignore air resistance, the penny falls with constant acceleration, and we can compute the time to reach the sidewalk and the velocity of the penny when it gets there.

* If we take air resistance into account, and drop the penny at its terminal velocity, it falls with constant velocity.

Now let's consider a third model that includes elements of the first two: let's assume that the acceleration of the penny is `a` until the penny reaches $29$ m/s, and then $0$ m/s$^2$ afterwards.  What is the total time for the penny to fall $381$ m?


You can break this question into three parts:

1. How long would the penny take to reach $29$ m/s with constant acceleration `a`.
2. How far would it fall during that time?
3. How long would it take to fall the remaining distance with constant velocity $29$ m/s?

Suggestion: Assign each intermediate result to a variable with a meaningful name.  And assign units to all quantities!

````python
# Answer Exercise 6 here
a = 9.8 * meter / second**2
h = 381 * meter
import numpy as np
import matplotlib.pyplot as plt

g = 9.8
distance = 18.44
h_release = 2.0
h_target = 0.75
v_initial = 40

angles = np.linspace(-10, 10, 100)
final_heights = []

for angle_deg in angles:
    angle_rad = np.radians(angle_deg)
    vx = v_initial * np.cos(angle_rad)
    vy = v_initial * np.sin(angle_rad)
    t_flight = distance / vx
    y_final = h_release + vy * t_flight - 0.5 * g * t_flight**2
    final_heights.append(y_final)

final_heights = np.array(final_heights)
idx = np.argmin(np.abs(final_heights - h_target))
best_angle = angles[idx]

plt.figure(figsize=(10, 6))
plt.plot(angles, final_heights, linewidth=2)
plt.axhline(y=0, linewidth=1)
plt.axhline(y=h_target, linestyle='--', linewidth=2)
plt.axvline(x=best_angle, linestyle='--', linewidth=2)
plt.xlabel('Initial Angle (degrees)')
plt.ylabel('Final Height at Home Plate (m)')
plt.title('Final Height vs Initial Angle')
plt.grid(True, alpha=0.3)
plt.show()

angle_rad = np.radians(best_angle)
vx = v_initial * np.cos(angle_rad)
vy = v_initial * np.sin(angle_rad)
t = np.linspace(0, distance / vx, 100)
x_traj = vx * t
y_traj = h_release + vy * t - 0.5 * g * t**2

plt.figure(figsize=(12, 6))
plt.plot(x_traj, y_traj, linewidth=2)
plt.scatter([0], [h_release], s=100)
plt.scatter([distance], [h_target], s=100)
plt.axhline(y=0, linewidth=1)
plt.axhline(y=h_target, linestyle='--')
plt.xlim(-1, distance + 1)
plt.ylim(-0.5, h_release + 1)
plt.xlabel('Distance (m)')
plt.ylabel('Height (m)')
plt.title(f'Baseball Pitch Trajectory (Initial Angle: {best_angle:.2f}°)')
plt.grid(True, alpha=0.3)
plt.show()

print("Analysis Results:")
print("=================")
print(f"Initial velocity: {v_initial} m/s (~{v_initial * 2.237:.1f} mph)")
print(f"Distance to plate: {distance} m")
print(f"Release height: {h_release} m")
print(f"Target height: {h_target} m")
print(f"\nRequired initial angle: {best_angle:.2f}°")

if best_angle > 0:
    print(f"\n✓ The ball is thrown UPWARD at {best_angle:.2f}° above horizontal")
else:
    print(f"\n✓ The ball is thrown DOWNWARD at {abs(best_angle):.2f}° below horizontal")
````

# Exercise 8 - [5 pt]

Suppose I run a 10K race in 44:52 (44 minutes and 52 seconds).  What is my average pace in minutes per mile?

````python
mile = units.mile
kilometer = units.kilometer
minute = units.minute
````

````python
race_time = 44 * minute + 52 * second
race_distance = 10 * kilometer

pace = race_time / race_distance
pace_per_mile = pace.to(minute / mile)

total_minutes_per_mile = pace_per_mile.magnitude
minutes_part = int(total_minutes_per_mile)
seconds_part = int((total_minutes_per_mile - minutes_part) * 60)

print(f"Race time: {race_time}")
print(f"Race distance: {race_distance}")
print(f"Pace: {pace}")
print(f"Average pace: {pace_per_mile}")
print(f"Average pace: {pace_per_mile.magnitude:.2f} minutes per mile")
print(f"\nAverage pace: {minutes_part}:{seconds_part:02d} per mile")
````
